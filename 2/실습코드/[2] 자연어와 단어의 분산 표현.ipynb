### 2.3.1 파이썬으로 말뭉치 전처리하기

```python
# 예시문장

text = 'You say goodbye and I say hello.'
```

```python
text = text.lower() # 모든 문자를 소문자로 변환
text = text.replace('.', ' .') # 마침표 고려
text
```

```python
words = text.split(' ') # 공백을 기준으로 변환
words
```

```python
# 딕셔너리를 이용하여 단어와 ID 짝지어주기

word_to_id = {} # 단어 -> ID
id_to_word = {} # ID -> 단어


for word in words:
    
    if word not in word_to_id: # 단어가 word_to_id에 들어있지 않으면
        
        new_id = len(word_to_id) # 새로운 id 추가
        word_to_id[word] = new_id # id 딕셔너리에 숫자 추가
        id_to_word[new_id] = word # 단어 딕셔너리에 단어 추가
```

```python
# 단어목록을 ID목록으로 변환

import numpy as np
corpus = [word_to_id[w] for w in words] # 단어 -> id
corpus = np.array(corpus)
corpus
```

```python
# 하나의 함수로 구현

def preprocess(text):
    text = text.lower()
    text = text.replace('.', ' .')
    words = text.split(' ')

    word_to_id = {}
    id_to_word = {}
    
    for word in words:
        
        if word not in word_to_id:
            new_id = len(word_to_id)
            word_to_id[word] = new_id
            id_to_word[new_id] = word

    corpus = np.array([word_to_id[w] for w in words])

    return corpus, word_to_id, id_to_word
```

### 2.3.4 동시발생 행렬

**통계 기반 기법**: 어떤 단어에 주목했을 때 그 주변에 어떤 단어가 몇번이나 등장하는지 세어 집계

#### 위 방법을 모든 단어에 적용하여 **동시발생행렬 생성**

```python
Image("image/fig 2-7.png")
```

```python
# 말뭉치로부터 동시발생 행렬을 만들어주는 함수 구현

def create_co_matrix(corpus, vocab_size, window_size=1):

    corpus_size = len(corpus)
    co_matrix = np.zeros((vocab_size, vocab_size), dtype=np.int32)

    for idx, word_id in enumerate(corpus):
        
        for i in range(1, window_size + 1):
            
            left_idx = idx - i
            right_idx = idx + i

            if left_idx >= 0:
                left_word_id = corpus[left_idx]
                co_matrix[word_id, left_word_id] += 1

            if right_idx < corpus_size:
                right_word_id = corpus[right_idx]
                co_matrix[word_id, right_word_id] += 1

    return co_matrix
```

### 2.3.5 벡터 간 유사도

단어 벡터의 유사도를 나타낼 때는 **코사인 유사도**를 자주 이용!

```python
Image("image/e 2-1.png")
```

```python
# 함수로 구현

def cos_similarity(x, y, eps=1e-8):

    nx = x / (np.sqrt(np.sum(x ** 2)) + eps) # 0으로 나누게 되는 오류를 범하지 않기 위해 eps 더해주기
    ny = y / (np.sqrt(np.sum(y ** 2)) + eps) 
    
    return np.dot(nx, ny)
```

#### 'you'와 'i'의 유사도

```python
text = 'You say goodbye and I say hello.'
corpus, word_to_id, id_to_word = preprocess(text)
vocab_size = len(word_to_id)
C = create_co_matrix(corpus, vocab_size)

c0 = C[word_to_id['you']]  # "you"의 단어 벡터
c1 = C[word_to_id['i']]    # "i"의 단어 벡터
print(cos_similarity(c0, c1))
```

### 2.3.6 유사 단어의 랭킹 표시

어떤 단어가 검색어로 주어지면 그 검색어와 비슷한 단어를 유사도 순으로 출력해보자

```python
def most_similar(query, word_to_id, id_to_word, word_matrix, top=5):

    if query not in word_to_id:
        print('%s(을)를 찾을 수 없습니다.' % query)
        return

    print('\n[query] ' + query)
    query_id = word_to_id[query]
    query_vec = word_matrix[query_id]

    # 코사인 유사도 계산
    vocab_size = len(id_to_word)

    similarity = np.zeros(vocab_size)
    
    for i in range(vocab_size):
        similarity[i] = cos_similarity(word_matrix[i], query_vec)

    # 코사인 유사도를 기준으로 내림차순으로 출력
    count = 0
    for i in (-1 * similarity).argsort():
        if id_to_word[i] == query:
            continue
        print(' %s: %s' % (id_to_word[i], similarity[i]))

        count += 1
        if count >= top:
            return
```

```python
text = 'You say goodbye and I say hello.'
corpus, word_to_id, id_to_word = preprocess(text)
vocab_size = len(word_to_id)
C = create_co_matrix(corpus, vocab_size)

most_similar('you', word_to_id, id_to_word, C, top=5)
```

## 2.4 통계 기반 기법 개선하기

### 2.4.1 상호정보량

- 단순히 동시발생 횟수가 잦다고 해서 큰 관련성을 갖는다고 할 수 없음 (ex. the car)
- 이 문제를 해결하기 위해 **점별 상호정보량(PMI)** 라는 척도 사용

#### PMI

```python
Image("image/e 2-3.png")
```

- **N**: 말뭉치에 포함된 단어 수, **P(x)**: x가 일어날 확률, **P(y)**: y가 일어날 확률, **P(x,y)**: x,y가 동시에 일어날 확률
- **C**: 동시발생 행렬, **C(x)**: x의 등장 횟수, **C(y)**: y의 등장 횟수, **C(x,y)**: x,y가 동시뱔생하는 횟수
- 각각의 단어가 등장하는 횟수로 나눠줌으로써 빈도 반영 -> 위 문제 해결 가능
- PMI 값이 높을 수록 관련성이 높음!
- 실제로 구현할 때는 **양의 상호정보량 (PPMI)** 사용

```python
Image("image/e 2-6.png")
```

```python
def ppmi(C, verbose=False, eps = 1e-8):
    
    M = np.zeros_like(C, dtype=np.float32)
    N = np.sum(C)
    S = np.sum(C, axis=0)
    total = C.shape[0] * C.shape[1]
    cnt = 0

    for i in range(C.shape[0]):
        
        for j in range(C.shape[1]):
            
            pmi = np.log2(C[i, j] * N / (S[j]*S[i]) + eps)
            M[i, j] = max(0, pmi)

            if verbose:
                cnt += 1
                if cnt % (total//100) == 0:
                    print('%.1f%% 완료' % (100*cnt/total))
    return M
```

```python
text = 'You say goodbye and I say hello.'
corpus, word_to_id, id_to_word = preprocess(text)
vocab_size = len(word_to_id)
C = create_co_matrix(corpus, vocab_size)
W = ppmi(C)

np.set_printoptions(precision=3)  # 유효 자릿수를 세 자리로 표시
print('동시발생 행렬')
print(C)
print('-'*50)
print('PPMI')
print(W)
```

### 2.4.2 차원 감소

'중요한 정보'는 최대한 유지하면서 차원을 감소시키는 것이 포인트


#### 특잇값 분해 (SVD)


```python
Image("image/e 2-7.png")
```

- 임의의 학렬을 세 행렬의 곱으로 분해
- 행렬 S에서 특잇값이 작다면 중요도가 낮다는 뜻이므로, 행렬 U에서 여분의 열벡터를 깎아내어 원래의 행렬 근사 가능

### 2.4.3 SVD에 의한 차원 감소

```python
text = 'You say goodbye and I say hello.'
corpus, word_to_id, id_to_word = preprocess(text)
vocab_size = len(id_to_word)
C = create_co_matrix(corpus, vocab_size, window_size=1)
W = ppmi(C)

# SVD
U, S, V = np.linalg.svd(W)

np.set_printoptions(precision=3)  # 유효 자릿수를 세 자리로 표시
print(C[0])
print(W[0])
print(U[0])
```

**'U[0]'(밀집벡터)를 2차원 벡터로 줄이려면 단순히 처음 두 원소를 꺼내면 됨! -> 'U[0, :2]'**
